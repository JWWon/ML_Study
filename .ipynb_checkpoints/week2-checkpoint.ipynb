```python
import numpy as np # Linear Algebra => Math
import pandas as pd # Data Frame - Data Format
from sklearn import linear_model # Machine Learning Packages
```

## Data Load

```python
data = pd.read_csv('hotel_energy.csv') # Data Frame
# data.iloc[:10, :] # iloc: index_location, :10 => 행 갯수 | : => 열

# Predict energy consumstion
x = data.iloc[:, 2:] # Data Frame
y = data.iloc[:, 1] # Data Frame / Series
```

## Model

```python
model = linear_model.LinearRegression()
model.fit(x, y)
coef = model.coef_
```

```python
pred_y = [0]*len(data.index)
for hotel in range(len(data.index)):
    for i in range(len(coef)):
        pred_y[hotel] += coef[i]*x.iloc[hotel, i]
```

```python
y - pred_y
```
