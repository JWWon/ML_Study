# Logistic

```python
import pandas as pd
```

```python
df = pd.read_csv('binary.csv')
```

```python
df # admit(합격여부)을 예측
# classification
# rank를 범주형으로 분석하는 것이 과제! 가장 대표적인 방법 찾아서 해보기
# 1. data를 변형
# 2. sklearn에서 패키지 검색
```

```python
x = df.iloc[:, 1:]
y = df.iloc[:, 0] # df.['admit']
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = \
    train_test_split(x, y, test_size = 0.3, random_state = 33)
```

## Scaling

```python
# scaled_train_x = (train_x - train_x.min())/(train_x.max() - train_x.min())
```

```python
from sklearn import preprocessing

scaler = preprocessing.StandardScaler().fit(train_x)
scaled_train_x = scaler.transform(train_x)
scaled_test_x = scaler.transform(test_x)
```

```python
scaled_train_x
```

## Modeling

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(fit_intercept=True, max_iter=1000)
```

```python
model.fit(scaled_train_x, train_y)
```

```python
model.coef_
model.intercept_
```

```python
model.score(scaled_test_x, test_y)
```
