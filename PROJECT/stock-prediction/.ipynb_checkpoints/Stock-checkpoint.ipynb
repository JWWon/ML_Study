## SP500 Prediction Model
- predict future value of 1 min later

```python
import tensorflow as tf
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
# Import Data
data = pd.read_csv('data_stocks.csv')
# Drop date
data = data.drop(['DATE'], 1)

# Dimension of dataset
n = data.shape[0]
p = data.shape[1]

# Data -> numpy array
data = data.values
# print SP500(Y value)
plt.plot(data[:, 0])
plt.show()

# Split training & test data
train_start = 0
train_end = int(np.floor(0.8*n))
test_start = train_end + 1
test_end = n

data_train = data[np.arange(train_start, train_end), :]
data_test = data[np.arange(test_start, test_end), :]

# TODO: if possible, should do nomalization

X_train = data_train[:, 1:]
y_train = data_train[:, 0]
X_test = data_test[:, 1:]
y_test = data_test[:, 0]
```

```python
# Initializers
# sigma = 1
# weight_initializer = tf.variance_scaling_initializer(mode="fan_avg", distribution="uniform", scale=sigma)
# bias_initializer = tf.zeros_initializer()

# Model architecture params
n_stocks = X_train.shape[1]
n_1_num = 1024
n_2_num = 512
n_3_num = 256
n_4_num = 128
n_target = 1
```

```python
# feedforward network

# Define Train Model
X = tf.placeholder(dtype=tf.float32, shape=[None, n_stocks])
Y = tf.placeholder(dtype=tf.float32, shape=[None])

# Layer 1
W1 = tf.Variable(tf.random_normal([n_stocks, n_1_num]), name='weight1')
b1 = tf.Variable(tf.random_normal([n_1_num]), name='bias1')
L1 = tf.nn.relu(tf.add(tf.matmul(X, W1), b1))
# Layer 2
W2 = tf.Variable(tf.random_normal([n_1_num, n_2_num]), name='weight2')
b2 = tf.Variable(tf.random_normal([n_2_num]), name='bias2')
L2 = tf.nn.relu(tf.add(tf.matmul(L1, W2), b2))
# Layer 3
W3 = tf.Variable(tf.random_normal([n_2_num, n_3_num]), name='weight3')
b3 = tf.Variable(tf.random_normal([n_3_num]), name='bias3')
L3 = tf.nn.relu(tf.add(tf.matmul(L2, W3), b3))
# Layer 4
W4 = tf.Variable(tf.random_normal([n_3_num, n_4_num]), name='weight4')
b4 = tf.Variable(tf.random_normal([n_4_num]), name='bias4')
L4 = tf.nn.relu(tf.add(tf.matmul(L3, W4), b4))
# Output
W_out = tf.Variable(tf.random_normal([n_4_num, n_target]), name='weight_out')
b_out = tf.Variable(tf.random_normal([n_target]), name='bias_out')
L_out = tf.transpose(tf.add(tf.matmul(L4, W_out), b_out))

# Cost function
cost = tf.reduce_mean(tf.squared_difference(L_out, Y))

# Optimizer
optimizer = tf.train.AdamOptimizer().minimize(cost)
```

```python
# Num of epochs & batch size
epochs = 10
batch_size = 256

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    # Setup interactive plot
    plt.ion()
    figure = plt.figure()
    ax1 = figure.add_subplot(111)
    line1, = ax1.plot(y_test)
    line2, = ax1.plot(y_test*0.5)
    plt.show()

    for e in range(epochs):
        print('epoch process: ' + str(e))
        # Shuffle training
        shuffle_indices = np.random.permutation(np.arange(len(y_train)))
        X_train = X_train[shuffle_indices]
        Y_train = y_train[shuffle_indices]
        
        # Minibatch training
        for i in range(0, len(y_train) // batch_size):
            start = i * batch_size
            batch_x = X_train[start: start + batch_size]
            batch_y = y_train[start: start + batch_size]
            # Run optimizer with batch
            sess.run(optimizer, feed_dict={X: batch_x, Y: batch_y})
            
            # Show progress
            if np.mod(i, 5) == 0:
                # Prediction
                prediction = sess.run(L_out, feed_dict={X: X_test})
                line2.set_ydata(prediction)
                plt.title('Epoch ' + str(e) + ', Batch ' + str(i))
                file_name = 'img/epoch_' + str(e) + '_batch_' + str(i) + '.png'
                plt.savefig(file_name)
                plt.pause(0.01)
                
    # Print final MSE after Training
    cost_final = sess.run(cost, feed_dict={X: X_test, Y: y_test})
    print(cost_final)
```
